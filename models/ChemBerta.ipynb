```python
data = pd.read_excel('/content/drive/MyDrive/atomick_hack_dataset_35000.xlsx', usecols=[2, 3])
```

```python
data.head()
```

```python
data = data.drop_duplicates()
```

```python
Q1 = data.IC50.quantile(q=.25)
Q3 = data.IC50.quantile(q=.75)
IQR = Q3 - Q1
```

```python
data = data[~((data.IC50 < (Q1-1.5*IQR)) | (data.IC50 > (Q3+1.5*IQR)))] #убираем данные, который не попадают в межквартильный размах
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import logging
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
model = ClassificationModel('roberta', 'seyonec/PubChem10M_SMILES_BPE_396_250',
                            use_cuda=False,
                            args={'evaluate_each_epoch': True,
                                  'evaluate_during_training_verbose': True,
                                  'no_save': True,
                                  'num_train_epochs': 10,
                                  'auto_weights': True,
                                  'regression': True,
                                  }) # You can set class weights by using the optional weight argument
```

```python
!wandb login
```

```python
train_df = data[:30000]
valid_df = data[30000:]
```

```python
train_df['label'] = train_df['IC50']
valid_df['label'] = valid_df['IC50']
```

```python
train_df.drop('IC50', axis=1, inplace=True)
```

```python
valid_df.drop('IC50', axis=1, inplace=True)
```
